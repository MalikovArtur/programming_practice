```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = [3, 5, 2, 10, 11]
b = [4, 2, 8, 8, 10]
```

```python
import numpy as np
a = np.array(a)
```

```python
a
```

```python
eps = np.array([0.5 for i in range(len(a))])
```

```python
eps
```

```python
plt.figure(figsize=(20, 10))
plt.plot(a, color='blue', linewidth=5)
plt.plot(a + eps, linestyle='--', color='red')
plt.plot(a - eps, linestyle='--', color='red')
plt.fill_between(range(len(a)), a - eps, a + eps, facecolor='green', alpha=0.1)
plt.title('Рэндомный график', fontsize=16)
plt.grid(True)
plt.xlabel('a', fontsize=18)
plt.ylabel('b', fontsize=16)
plt.show()
```
